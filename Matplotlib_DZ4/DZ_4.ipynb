**_Коментарий_**
Условия задачи 3 написаны давольно размыто, поэтому не совсем понятно, что требуется выполнить.
Построила таблицу корелляций, взяла понятные мне признаки с разным коэффициентом кореляции (sqft_living15 с коэфицентом кореляции 0,6 не рассматривала, т.к. не понятен его смысл) и построила разные типы диаграмм по ним.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot= True, linewidths= 0.5);
```

### Вывод:
Из матрицы кореляции видно, что на цену дома влияют несколько факторов:
- жилая площадь и оценка имеют самое высокое влияние на цену дома, с коэффициентом кореляции 0,7;
- площадь дома без цоколя и колличество ванных комнат заметно влияют на цену дома, коэффициент кореляции 0,6 и 0,5 соответственно;
- вид умеренно влияет на цену дома, коэффициент кореляции 0,4.

```python
plt.figure(figsize= (15,6))
sns.lineplot(y= df['sqft_living'], x= df['price'])
plt.title('Зависимость цены дома от жилой площади');
```

### Вывод:
Прослеживается линейная зависимость цены от жилой площади дома.
С единичным выбросом, который говорит о том, что цена низкая, а площадь большая, следовательно с домом что-то не так.

```python
plt.figure(figsize= (8,6))
sns.scatterplot(x= df['grade'], y= df['price'])
plt.title('Зависимость цены дома от оценки');
```

### Вывод:
Линейная зависимость цены от оценки дома. С выбросом, свидетельствующем о том, что дом был продан дороже своей оценки.

```python
plt.figure(figsize= (15,8))
sns.jointplot(x= df['sqft_above'], y= df['price'])
plt.title('Зависимость цены дома от площади без цоколя');
```

### Вывод:
Зависимость с умеренным ростом, цены от площади дома без цоколя.
С еденичным выбросом, как в первом графике, большая площадь с низкой ценой.
Плотность точен на графике, указывает на то, что наибольшее колличество домов с площадью 1100 по цене 500000.

```python
plt.figure(figsize= (15,6))
sns.boxplot(y= df['price'], x= df['bathrooms'], showfliers= False)
plt.title('Зависимость цены дома от кол-ва ванных комнат');
```

### Вывод:
Заметное влияние количества ванных комнат на цену дома, чем больше ванных комнат тем выше цена на дом.
Есть выброс на 6,5 ванных комнат, цена явно ниже ожидаемой.


```python
plt.figure(figsize= (8,6))
sns.violinplot(x= df['view'], y= df['price'])
plt.title('Зависимость цены дома от вида');
```

### Вывод:
Умеренное влияние вида на цену дома.
